```python
y_classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y = train[y_classes].values
```

```python
from keras.models import Model
from keras.layers import Dense, Embedding, Input
from keras.layers import LSTM, Bidirectional, GlobalMaxPool1D, Dropout
from keras.preprocessing import text, sequence
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
tokenizer = text.Tokenizer(num_words=20000)
tokenizer.fit_on_texts(norm_train_comments)
list_tokenized_train = tokenizer.texts_to_sequences(norm_train_comments)
list_tokenized_test = tokenizer.texts_to_sequences(norm_test_comments)
X_t = sequence.pad_sequences(list_tokenized_train, maxlen=100)
X_te = sequence.pad_sequences(list_tokenized_test, maxlen=100)
```

```python
X_te.shape
```

```python
def get_model():
    embed_size = 128
    inp = Input(shape=(100, ))
    x = Embedding(20000, embed_size)(inp)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.1)(x)
    x = Dense(50, activation="relu")(x)
    x = Dropout(0.1)(x)
    x = Dense(6, activation="sigmoid")(x)
    model = Model(inputs=inp, outputs=x)
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    return model
```

```python
model = get_model()
batch_size = 100
model.fit(X_t, y, epochs=5, batch_size=batch_size, 
          shuffle=True, validation_split=0.1, verbose=1)
```
